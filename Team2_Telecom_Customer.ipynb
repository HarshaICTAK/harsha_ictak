```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
data.head()
```

# Churn count with respect to Gender

```python
sns.countplot(x=data['Churn'], hue=data['gender'])
plt.title('Churn count with respect to gender')
```

# Number of Female Senior Citizens

```python
data.groupby(['SeniorCitizen','gender'])['gender'].count()
```

```python
sns.countplot(x=data['gender'], hue=data['SeniorCitizen'])
plt.title('Number of Senior Citizens')
```

```python
#From the above information, we can infer that 568 females are Senior citizens.
```

# Plot of Tenure VS Total Charges

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=data, x='tenure', y='TotalCharges')
plt.title('Tenure VS Total Charges')
```

# Contract preferred by Senior Citizens

```python
sns.countplot(x=data['Contract'], hue=data['SeniorCitizen'])
plt.title('Contract with respect to Senior Citizen')
```

```python
#The month-to-month contract is the most preferred by Senior Citizens.
```

# Comments on Payment Method

```python
plt.figure(figsize=(10,8))
sns.countplot(x='PaymentMethod', hue='gender', data=data)
plt.title('Payment Method with respect to gender')
```

```python
#While, men tend to prefer electronic check,credit card and mailed check methods for payment, women tend to prefer bank transfer as method of payment. 
```
