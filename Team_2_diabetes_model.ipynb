```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. Create a Count plot of Outcome 
# 2. Analyse the effect of age on Outcome using Boxplot
# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function 
# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome. 
# 5. Create a heatmap on Correlation of features in the data.
```

```python
data = pd.read_csv('diabetes_model.csv')
```

```python
data.head()
```

# Countplot of Outcome

```python
sns.countplot(data['Outcome'])
plt.grid()
```

# Effect of Age on Outcome

```python
plt.figure(figsize=(10,8))
sns.boxplot(x='Outcome', y='Age', data=data)
plt.title('Effect of Age on Outcome')
```

# Diabetes Probability

```python
plt.figure(figsize=(10,8))
sns.stripplot(x=data['Outcome'], y=data['DiabetesPedigreeFunction'])
plt.title('Probability of Diabetes in family')
plt.xlabel('Outcome')
plt.ylabel('DiabetesPedigreeFunction')
```

# Insulin and Glucose

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Insulin', y='Glucose', hue='Outcome', data=data)
plt.title('Relationship between Insulin & Glucose')
```

# Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True)
```
