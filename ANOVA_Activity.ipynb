# 1. Explain the dataset

The data shows the sales of a company based on the different forms of advertising methods used by the manager. The amount spent for advertising using TV, Radio, and Newspaper and the Sales achieved respectively are given.

```python
plt.figure(figsize=(12,10))

plt.subplot(3,1,1)
sns.histplot(data['TV'], color='green')
plt.title('Amount spent for TV')

plt.subplot(3,1,2)
sns.histplot(data['Radio'], color='red')
plt.title('Amount spent for Radio')

plt.subplot(3,1,3)
sns.histplot(data['Newspaper'], color='blue')
plt.title('Amount spent for Newspaper')

plt.tight_layout()
```

```python
plt.figure(figsize=(12,10))

plt.subplot(3,1,1)
sns.scatterplot(data['TV'],data['Sales'], color='green')
plt.title('TV vs Sales')

plt.subplot(3,1,2)
sns.scatterplot(data['Radio'], data['Sales'],color='red')
plt.title('Radio vs Sales')

plt.subplot(3,1,3)
sns.scatterplot(data['Newspaper'],data['Sales'], color='blue')
plt.title('Newspaper VS Sales')

plt.tight_layout()
```

# 3. Testing of means of expenses for advertisements

We can consider the hypotheses to be:
    
    Null Hypothesis H0: Same amount was spent for all three advertisements
    Alternative Hypothesis H1: Atleast for one advertisement different amount was spent as compared to the rest

```python
st.f_oneway(data['TV'],data['Radio'],data['Newspaper'])
```

The F-statistic is 358.85 and the corresponding p-value is 4.552931539744962e-103. 

Since p-value is very small as compared to 0.05, we can reject the null hypothesis.

Hence, we can conclude that atleast one form of advertisement had different amount spent by the manager.
